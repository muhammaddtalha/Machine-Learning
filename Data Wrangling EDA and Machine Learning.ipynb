### Section 2: Deal with missing data

```python
#If the missing values were close to or greater than 70%, we may have deleted this column.
#As there are only 20% of the values are missing therefore, Now we have 2 choices, droping rows with null values or replace them with some technique.
#Median of a distribution is robust to the outliers, therefore, we are replaceing the missing values with the median values.
#We are moving with the replacement of null/missing values with median.

df_w.fillna(df_w.median(), inplace = True)
```

```python
#Again checking the null values in the dataframe.
#there are 2 null values in the num-of-doors, because it was categorical variable, we have replace it with other technique.
df_w.isna().sum()
```

```python
#Replacing missing values in num-of-doors column with its mode value
df_w['num-of-doors'].fillna(df_w['num-of-doors'].mode()[0], inplace = True)
```

# Machine Learning

Loading data

```python
data_c = pd.read_csv('classification.csv')
df_c = data_c.copy()
df_c.head()
```

```python
df_c.shape
```

## Decision Tree

1. Create dependent and independent variables
2. Split data into train and test with test size of 20% and random state = 25
3. Import decision tree classifier and train it with default settings and random state = 25
4. Get predictions on the trained model
5. Get classification report for the model

**Task 1: Creating dependent and independent variables**

*Note: Target_Flag is dependent variable*
- df_c.TARGET_FLAG

```python
# write code here
X = df_c.drop('TARGET_FLAG', axis = 1)
Y = df_c.TARGET_FLAG
```

```python
print(X.shape)
print(Y.shape)
```

**Task 2: Splitting data into train and test with test size of 20% and random state = 25**

```python
from sklearn.model_selection import train_test_split
# Write the code below to split the data
xtrain, xtest, ytrain, ytest = train_test_split(X,Y,test_size = 0.2, random_state=25)
```

```python
# Print the shapes of above split data
print(xtrain.shape, ytrain.shape)
print(xtest.shape, ytest.shape)
```

**Task 3: Importing decision tree classifier and training it with default settings and random state = 25**

```python
# Write the code to import decision tree from sklearn
from sklearn import tree
clf = tree.DecisionTreeClassifier(random_state=25)
```

```python
# Train the model
clf = clf.fit(xtrain, ytrain)
```

**Task 4: Get predictions**


```python
# Write the code to get predictions
pred = clf.predict(xtest)
```

**Task 5: Get Classification report**

```python
# Write the code here
from sklearn.metrics import classification_report
print(classification_report(ytest, pred))
```

## Section 2: Random Forest
